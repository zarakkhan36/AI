```python
# Zarak Khan
# CSC 362 AI
# 9/26/25
# Problem 3
# HW2 - Purpose: Implement DFS to check if goal is reachable

from collections import deque

def DFS(graph, start, goal):

    # DFS stack
    stack = deque([start])
    visited = set([start])

    while stack:
        vertex = stack.pop()  # Take last added node

        # If goal is reachable, return success
        if vertex == goal:
            return "SUCCESS"

        # Search neighbors of the current vertex
        # Use reversed() to enforce left-to-right ordering
        for neighbor in reversed(graph[vertex]):
            if neighbor not in visited:
                visited.add(neighbor)
                stack.append(neighbor)

    # If stack empties without finding goal
    return "FAILURE"


```
