# **Breadth-First Search (BFS)**

* To search through a graph, a search agent needs to **expand** nodes
* The nodes which can be expanded by the agent together form the **frontier**
* Expanding a node refers to marking the node as **expanded** or **visited** and adding its immediate neighbors to the frontier.
* In the following diagram, '0' is the initial node and belongs to the frontier.

* The search agent starts traversing from a selected node (source or starting node) and then traverses the graph layerwise thus exploring the neighbor nodes (nodes which are directly connected to source node). The agent then moves towards the next-level neighbor nodes.

image.png

* A graph may contain cycles, which can bring the agent the same node again while traversing the graph. To avoid processing of same node again, we use a boolean array which marks the node after it is processed. While visiting the nodes in the layer of a graph, store them in a manner such that you can traverse the corresponding child nodes in a similar order.

* To make this process easy, we use a **queue** to store the node and mark it as **visited** until all its neighbors (vertices that are directly connected to it) are marked.

* The queue follows the **First In First Out (FIFO**) queuing method, and therefore, the neigbors of the node will be visited in the order in which they were inserted in the queue i.e. the node that was inserted first will be visited first, and so on.

# **Deque in Python**

* **Deque (Doubly Ended Queue)** in Python is implemented using the module **collections** (https://docs.python.org/3/library/collections.html).

* **Deque** is preferred over a list in the cases where we need quicker append and pop operations from both the ends of the container, as deque provides an O(1) time complexity for **append** and **pop** operations as compared to a list that provides O(n) time complexity.

image.png


```python

# Zarak Khan
# CSC 362 AI
# 9/22/25
# Problem 1
# HW2 - Purpose: Implement BFS to find kth level friends for a given user

# Import deque
from collections import deque

def FindFriends(graph, user, k):

    # Keep track of visited nodes to avoid repetition
    visited = set([user])

    # Queue that stores the pair (node, level)
    queue = deque([(user, 0)])

    # List that collects kth level friends
    result = []

    # BFS loop
    while queue:
       # Take first element from queue
       current, level = queue.popleft()

       # If node is at level k, add it to result
       if level == k:
           result.append(current)

       # If not at level k, check the user's friends
       elif level < k:
           for neighbor in graph [current]:
               if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, level+1))

    # Ensure starting user (root) not included in result
    if user in result:
        result.remove(user)
    return result


```

# **Applications of BFS**

* **Crawlers in Search Engines**: Crawlers build an index using BFS. The idea is to start from the source page and follow all links from the source and keep doing the same. Depth First Traversal can also be used for crawlers, but the advantage of Breadth First Traversal is, the depth or levels of the built tree can be limited.

* **Social Networking Websites**: In social networks, we can find people within a given distance from a person using BFS.
